```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import cv2 as cv
```

```python
img = cv.imread('./images/input.jpg')
cv.imshow('Original Photo', img)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
print(type(img))
print(img.shape)
print("dimm: H x W X RGB")
print("image height: ", int(img.shape[0]), 'pixels')
print("image width: ", int(img.shape[1]), 'pixels')
print("image channels: ", int(img.shape[2]), 'channels')
```

```python
# open CV loads based on BGR compared to other libraries that load based on RGB
# plt.imshow(img)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
img2 = img
img2[200:300,300:500,:] = (0,0,255)
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
```
